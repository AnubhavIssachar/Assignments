**1. What is the difference between a function and a method in Python?**

Function: A standalone block of code that performs a specific task and can be called independently.

Method: A function that is associated with an object and is called on that object.


```python
#Function Example
def add(a, b):
    return a + b
result = add(3, 5)
print(result)
```

```python
#Method Example
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(3, 5)
print(result)
```

**5. What are iterators in Python and how do they differ from iterables?**

Iterable: An object that can be iterated over (e.g., lists, tuples). It has an __iter__() method.

Iterator: An object that represents a stream of data; it is produced from an iterable and has a __next__() method.

```python
#Iterable Example

my_list = [1, 2, 3]
```

```python
#Iterator Example
my_iter = iter(my_list)
print(next(my_iter))
```

**6. Explain the concept of generators in Python and how they are defined.**

Generators are a type of iterable that generate values on the fly using the yield keyword instead of returning a value.

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

counter = count_up_to(5)
for number in counter:
    print(number)  # Outputs: 1, 2, 3, 4, 5
```

**7. What are the advantages of using generators over regular functions?**

1-Memory Efficiency: Generators yield one item at a time, reducing memory usage.

2-Lazy Evaluation: Values are generated only when needed, which can lead to better performance in some cases.

3-State Retention: Generators retain their state between calls.

**8. What is a lambda function in Python and when is it typically used?**

A lambda function is an anonymous function defined with the lambda keyword. It can take any number of arguments but can only have one expression.

```python
add = lambda x, y: x + y
result = add(2, 3)
print(result)
```


**9. Explain the purpose and usage of the map() function in Python.**

The map() function applies a given function to all items in an iterable (like a list).

```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)
```

**10. What is the difference between map(), reduce(), and filter() functions in Python?**

1. map(): Applies a function to all items in an iterable.

Example: map(func, iterable)

2. reduce(): Applies a rolling computation to sequential pairs of values in a list (requires functools).

```python
from functools import reduce
result = reduce(lambda x, y: x + y, [1, 2, 3, 4])
print(result)
```

3.filter(): Filters items in an iterable based on a function that returns True or False.

```python
numbers = [1, 2, 3, 4]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)
```
