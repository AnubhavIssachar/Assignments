#3Compare and contrast mutable and immutable objects in Python with examples.

Mutable Objects:-

a)Definition- Objects that can be changed after creation
b)Examples- Lists, Dictionaries, Sets
c)Memory Allocation-Can modify in place, no new object created.
d)Methods-Have methods that modify the object (e.g., .append(), .remove())
e)Use Cases-When the data needs to be modified
f)Example Code:-

my_list = [1, 2, 3]
print("Original list:", my_list)

my_list.append(4)
print("Modified list:", my_list)

Immutable Objects:-

a)Definition-Objects that cannot be changed after creation.
b)Examples-Tuples, Strings, Frozensets
c)Memory Allocation-A new object is created for modifications.
d)Methods-Methods return new objects instead of modifying the existing one (e.g., .replace())
e)Use Cases	-When data integrity is important (e.g., using as keys in dictionaries).
f)Example Code:-

my_tuple = (1, 2, 3)
print("Original tuple:", my_tuple)

my_tuple[0] = 4  # This would raise a TypeError

new_tuple = my_tuple + (4,)
print("New tuple:", new_tuple)  # Output: (1, 2, 3, 4)

#6  How do conditional statements work in Python? Illustrate with examples.

The primary conditional statements are if, elif, and else

1) if Statement: Checks a condition and executes a block of code if the condition is true.

2)elif Statement: Stands for "else if" and checks another condition if the previous if statement was false.

3)else Statement: Executes a block of code if none of the previous conditions are true.


#Examples:-
1. if Statement
x = 10

if x > 5:
    print("x is greater than 5")

2. Using elif and else

x = 5

if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
